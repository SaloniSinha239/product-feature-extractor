```python
few_shot_prompt = """Extract the key product features mentioned in the following customer reviews. Focus on aspects like battery life, camera quality, screen resolution, performance, overheating, etc.

Review: "The battery lasts all day and the camera is amazing."
Features: battery life, camera quality

Review: "I love the screen resolution but the phone heats up quickly."
Features: screen resolution, overheating

Review: "Performance is smooth, but the battery drains fast."
Features: performance, battery life

Review: "Camera quality is top-notch, especially in low light."
Features: camera quality, low-light performance

Review: "{}"
Features:"""
```

```python
def extract_features(review):
    prompt = few_shot_prompt.format(review)
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)
    outputs = model.generate(**inputs, max_new_tokens=50)
    features = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return features.strip()
```

```python
results = []
for i, review in enumerate(reviews, 1):
    features = extract_features(review)
    results.append({"Review ID": i, "Review Text": review, "Extracted Features": features})
```

```python
output_file = "extracted_features.csv"
with open(output_file, mode="w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["Review ID", "Review Text", "Extracted Features"])
    writer.writeheader()
    writer.writerows(results)

print(f"Feature extraction complete. Results saved to {output_file}.")
```
